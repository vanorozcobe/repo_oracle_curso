#游댢 Normalizacion y Transformaci칩n de los datos

Como se tiene tanta informacion anidada es imposible que podamos leer la informacion es por esta razon que vamos a normalizar los datos para mayor facilidad de analisis y comprension

```python
#vamos a normalizar todas las columnas que tienen informacion dentro de la informacion
df_normalizado = pd.json_normalize(datos_clientes["customer"])
df_normalizado
```

```python
import pandas as pd

# Normalizar cada columna que contiene diccionarios
customer_df = pd.json_normalize(datos_clientes['customer'])
phone_df    = pd.json_normalize(datos_clientes['phone'])
internet_df = pd.json_normalize(datos_clientes['internet'])
account_df  = pd.json_normalize(datos_clientes['account'])

# Concatenar todo en un solo DataFrame para la mayor comprension de los datos
df_normalizado = pd.concat([datos_clientes[['customerID', 'Churn']],
                      customer_df, phone_df, internet_df, account_df], axis=1)

display(df_normalizado)
```

```python
display(df_normalizado)
```

Como ya tenemos toda la informacion correspondiente debemos verificar que tipo de datos es cada uno de los datos,  y cambiarla en caso de ser necesario.

```python
df_normalizado.info()
```

```python
tipo_datos = df_normalizado.info()
tipo_datos
```

```python
import numpy as np
import pandas as pd

# Reemplazar espacios en blanco por NaN
df_normalizado["Charges.Total"] = df_normalizado["Charges.Total"].replace(" ", np.nan)

# Convertir a float
df_normalizado["Charges.Total"] = df_normalizado["Charges.Total"].astype(float)

df_normalizado.info()


```

```python
#vamos a cambiar varios tipos de datos a la vez que aplican para ser booleanos

# Lista de columnas a convertir
convertir_bool = [
    "Partner", "Dependents", "PhoneService", "MultipleLines",
    "OnlineBackup", "DeviceProtection", "TechSupport",
    "StreamingTV", "StreamingMovies", "PaperlessBilling"
]

# Convertir a boolean: Yes=True, No=False, otros=NaN
df_normalizado[convertir_bool] = df_normalizado[convertir_bool].replace(
    {"Yes": True, "No": False, "No internet service": False, "No phone service": False}
).astype("boolean")

df_normalizado.info()
```

```python
#vamos a cambiar varios tipos de datos a la vez que aplican para ser strings

convertir_str = [
    "customerID",
    "Churn",
    "gender",
    "InternetService",
    "OnlineSecurity",
    "Contract",
    "PaymentMethod"
]

df_normalizado[convertir_str] = df_normalizado[convertir_str].astype("string")

df_normalizado.info()
```

#游늵 An치lisis

```python
display(df_normalizado)
```

```python
#Analisis de los datos suministrados

#Este analisis muestra el  % de clientes se quedaron y qu칠 % se fueron

df_normalizado["Churn"].value_counts(normalize=True) * 100
df_normalizado
```

```python
#Relacion entre tipo de contrato y churn
pd.crosstab(df_normalizado["Contract"], df_normalizado["Churn"], normalize="index") * 100

#en este analisis se vera los clientes con contrato a dos a침os los cuales casi no se van, mientras que los de mes a mes tienen mayor churn.
```

```python
# Analisis de la distribuci칩n del Churn

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.countplot(data=df_normalizado, x="Churn")
plt.title("Distribuci칩n de Churn")
plt.xlabel("쮺liente que se fue?")
plt.ylabel("Cantidad de clientes")
plt.show()
```

```python
#Churn por tipo de contrato


plt.figure(figsize=(6,4))
sns.countplot(data=df_normalizado, x="Contract", hue="Churn")
plt.title("Churn por tipo de contrato")
plt.xlabel("Tipo de contrato")
plt.ylabel("Cantidad de clientes")
plt.xticks(rotation=20)
plt.show()
```

```python
#Churn por m칠todo de pago

plt.figure(figsize=(6,6))
df_normalizado["PaymentMethod"].value_counts().plot.pie(
    autopct="%1.1f%%",
    startangle=90,
    counterclock=False
)
plt.ylabel("")  # quitar etiqueta
plt.title("Distribuci칩n de m칠todos de pago")
plt.show()
```

```python
#Seguridad comparada con  online - Abandono de clientes

plt.figure(figsize=(6,4))
sns.boxplot(
    data=df_normalizado,
    x="Churn",
    y="Charges.Monthly",
    palette={"Yes": "#E74C3C", "No": "#2ECC71"}  # rojo churn, verde no churn
)
plt.title("Cargos mensuales seg칰n Churn", fontsize=14)
plt.xlabel("Churn", fontsize=12)
plt.ylabel("Cargos mensuales", fontsize=12)
plt.show()
```
